```python
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

# Resultados del censo nacional de poblacion y vivienda

```python
etiquetas = ['Censados en hogares', 'LEA', 'Efectivamente censados', 'Personas omitidas', 'Estimación total']
valores = [43835324, 329093, 44164417, 4094077, 48258494]

plt.bar(etiquetas, valores, color='blue')

plt.title('Resultados del Censo Nacional de Población y Vivienda')
plt.xlabel('Categorías')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Número de personas')

# Agregar la cantidad en cada barra
for i, valor in enumerate(valores):
    plt.text(i, valor + 1000000, str(valor), ha='center', fontweight='bold')

plt.savefig('grafico_censo.png', dpi=300, bbox_inches='tight')
```

## Porcentaje y relación de mujeres

```python
porcentaje_mujeres = 51.2
proporcion_mujeres = 104.7

# Crear una figura con un solo eje
fig, ax = plt.subplots()

# Crear las barras
ax.barh(['Porcentaje de mujeres', 'Proporción de mujeres por cada 100 personas'], 
        [porcentaje_mujeres, proporcion_mujeres], 
        color=['blue', 'green'])

# Agregar los valores en cada barra
for i, v in enumerate([porcentaje_mujeres, proporcion_mujeres]):
    ax.text(v + 1, i, str(v), color='black', fontweight='bold')

# Etiquetas de los ejes y título
ax.set_xlabel('Porcentaje/Proporción')
ax.set_title('Relación entre el porcentaje de mujeres y la proporción de mujeres por cada 100 personas')

# Guardar el gráfico en formato png
plt.savefig('relacion_mujeres.png', dpi=300, bbox_inches='tight')

# Mostrar el gráfico
plt.show()
```

## Porcentaje y relación de hombres

```python
porcentaje_hombres = 48.8
proporcion_hombres = 95.5

# Crear una figura con un solo eje
fig, ax = plt.subplots(figsize=(8, 4))

# Crear las barras
ax.barh(['Porcentaje de hombres', 'Proporción de hombres por cada 100 mujeres'], 
        [porcentaje_hombres, proporcion_hombres], 
        color=['#0077b6', '#00b4d8'])

# Añadir texto encima de cada barra
for i, v in enumerate([porcentaje_hombres, proporcion_hombres]):
    ax.text(v + 0.5, i, str(v) + '%', color='#03045e', fontweight='bold', fontsize=14)

# Cambiar los colores de los ejes y de la figura
ax.spines['bottom'].set_color('#cccccc')
ax.spines['left'].set_color('#cccccc')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(axis='x', colors='#cccccc')
ax.tick_params(axis='y', colors='#cccccc')
ax.xaxis.label.set_color('#cccccc')
ax.yaxis.label.set_color('#cccccc')
ax.title.set_color('#cccccc')

# Etiquetas de los ejes y título
ax.set_xlabel('Porcentaje/Proporción', fontsize=14)
ax.set_title('Relación entre el porcentaje de hombres y la proporción de hombres por cada 100 mujeres', fontsize=18)

# Ajustar la posición de los ejes
ax.set_xlim([0, 110])
ax.invert_yaxis()

# Cambiar el fondo de la figura
fig.patch.set_facecolor('#f1f1f1')
ax.set_facecolor('#f1f1f1')

# Mostrar el gráfico
#plt.show()

plt.savefig('grafico_generos.png', dpi=300, bbox_inches='tight')
```

## Grandes grupos de edad

```python
grupos_edad = ['0 a 14 años', '15 a 59 años', 'Más de 59 años']
porcentajes = [22.6, 64.1, 13.3]

fig, ax = plt.subplots()

ax.bar(grupos_edad, porcentajes, color=['#0077b6', '#00b4d8', '#90e0ef'])
for i, v in enumerate(porcentajes):
    ax.text(i, v + 0.5, str(v) + '%', color='#03045e', fontweight='bold', fontsize=14, ha='center')

ax.spines['bottom'].set_color('#cccccc')
ax.spines['left'].set_color('#cccccc')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(axis='x', colors='#cccccc')
ax.tick_params(axis='y', colors='#cccccc')
ax.xaxis.label.set_color('#cccccc')
ax.yaxis.label.set_color('#cccccc')
ax.set_xlabel('Grupos de edad', fontsize=14)
ax.set_ylabel('Porcentaje', fontsize=14)
ax.set_title('Distribución de la población por grupos de edad', fontsize=18)

fig.set_size_inches(8, 6)
plt.savefig('grafico_edad.png', dpi=300, bbox_inches='tight')

plt.show()
```

## La población étnica en Total Nacional se autoreconoció como:

```python
# Datos
poblacion_etnica = [4.3, 6.8, 0.0, 88.8, 0.1, 0.0]
grupos_etnicos = ['Indígenas', 'Afrocolombianos', 'Rom-gitanos', 'Ningún grupo étnico', 'Raizales', 'Palenquero']

# Colores
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(10,6))

# Crear gráfico de barras horizontales
ax.barh(grupos_etnicos, poblacion_etnica, color=colores, height=0.8)

# Personalizar título y etiquetas de ejes
ax.set_title('Distribución de la población étnica en Colombia', fontsize=18)
ax.set_xlabel('Porcentaje (%)', fontsize=14)
ax.set_ylabel('Grupo étnico', fontsize=14)

# Cambiar tamaño de fuente de etiquetas de ejes
ax.tick_params(axis='both', which='major', labelsize=12)

# Añadir valores al final de las barras
for i, v in enumerate(poblacion_etnica):
    ax.text(v + 0.5, i, str(v)+'%', color='black', fontsize=12, va='center')

# Mostrar gráfico
plt.show()

# Guardar imagen en formato PNG
fig.savefig('diagrama_moderno.png', dpi=300, bbox_inches='tight')
```

## Total de viviendas en Colombia (2005-2018)

```python
# Datos
viviendas = [
    10570899, 
    14243223
]

years = [
    2005, 
    2018
]

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(8, 6))

# Crear gráfico de barras
ax.bar(years, viviendas, color=['#1f77b4', '#ff7f0e'])

# Personalizar título y etiquetas
ax.set_title('Total de viviendas en Colombia (2005-2018)', fontsize=18)
ax.set_ylabel('Número de viviendas', fontsize=14)
ax.set_xticks(years)
ax.ticklabel_format(style='plain', axis='y') # Mostrar números enteros en el eje y

# Mostrar gráfico
plt.show()

# Guardar imagen en formato PNG
fig.savefig('total_viviendas.png', dpi=300, bbox_inches='tight')
```

## Distribución de edades

```python


# Datos
edades = ['0-14', '15-65', '≥65']
frecuencias = [10906420, 32912293, 4391523]

# Crear el histograma
plt.hist(edades, weights=frecuencias, bins=len(edades), rwidth=1)

# Configurar el gráfico
plt.title('Distribución de edades')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.grid(axis='y')

# Guardar el gráfico en formato png
plt.savefig('histograma_edades.png')

# Mostrar el gráfico
plt.show()
```

## Análisis estadistico

```python
# Datos
servicios = ['E. Electrica', 'Acueducto', 'Alcantarillado', 'Gas', 'Aseo', 'Internet']
valores = [46472930, 41695339, 36966006, 32236674, 39378931, 20944186]

# Media
media = statistics.mean(valores)
print("Media:", media)

# Mediana
mediana = statistics.median(valores)
print("Mediana:", mediana)

# Desviación estándar
desviacion = statistics.stdev(valores)
print("Desviación estándar:", desviacion)

# Varianza
varianza = statistics.variance(valores)
print("Varianza:", varianza)
```
